```python
#Luego de observar que Dinamarca tiene el mayor indice de Felicidad y por el contrario Afganistan tiene el indice mas bajo de felicadad, se busca crear una tabla donde se pueda comparar todos los datos de estos dos paises
```

```python
# Filtrar el DataFrame para obtener solo las filas de Dinamarca y Afganistán
df_filtrado = data_felicidad[data_felicidad['pais'].isin(['Dinamarca', 'Afghanistán'])]
```

```python
df_filtrado
```

```python
resumen = df_filtrado.groupby('pais').agg(['mean', 'median'])
```

```python
resumen
```

```python
#se puede observar la diferencia significativa de los dos Paises en relacion al pib,el soporte social, la expectativa de vida, la calidad de democracia y calidad de entrega que supera Dinamarca a Afghanistan inclusive siendo estos ultimos indices negativos para Afghanis
```
